```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
A = np.array([[2,3,5],[3,1,4],[3,4,12]])
```

```python
origin = np.zeros(3)
```

```python
def three_cols(arr):
    return arr[:, 0], arr[:, 1], arr[:, 2]
```

```python
v1, v2, v3 = three_cols(A)
```

```python
def create_plot(arr, korigin=origin):
	fig = plt.figure(figsize=(10, 10))
	ax = fig.add_subplot(111, projection='3d')

	v1, v2, v3 = three_cols(arr)

	# Plot the vectors
	ax.quiver(*origin, *v1, color='r', label='Column 1')
	ax.quiver(*origin, *v2, color='g', label='Column 2')
	ax.quiver(*origin, *v3, color='b', label='Column 3')

	# Set labels and title
	ax.set_xlabel('X')
	ax.set_ylabel('Y')
	ax.set_zlabel('Z')
	ax.set_title('3x3 Matrix Columns as Vectors in 3D Space')
	# Set axis limits
	ax.set_xlim([-15, 15])
	ax.set_ylim([-15, 15])
	ax.set_zlim([-15, 15])

	# Add a legend
	ax.legend()

	# Show the plot
	plt.show()
```

```python
create_plot(A)
```

```python
import numpy as np

A = np.array([[2, 1, 0],  # column a
              [0, 1, -1], # column b
              [1, 0, 2]]) # column c

# Original unit vectors
x_unit = np.array([1, 0, 0])
y_unit = np.array([0, 1, 0])
z_unit = np.array([0, 0, 1])

# Where these unit vectors end up after transformation
x_transformed = A @ x_unit  # same as A[:, 0]
y_transformed = A @ y_unit  # same as A[:, 1]
z_transformed = A @ z_unit  # same as A[:, 2]

print("Transformed x-unit:", x_transformed)
print("Transformed y-unit:", y_transformed)
print("Transformed z-unit:", z_transformed)
```

```python
import numpy as np

# 3x5 matrix
A = np.array([[1, 2, 0, -1, 3],
              [0, 1, 2, 1, -1],
              [2, 0, -1, 1, 0]])

# A point in 5D space
v = np.array([1, 2, 3, 4, 5])

# Transform the 5D point to 3D
result = A @ v

print("Original 5D point:", v)
print("Transformed 3D point:", result)

# Let's see where each 5D basis vector ends up in 3D
for i in range(5):
    basis_5d = np.zeros(5)
    basis_5d[i] = 1
    transformed = A @ basis_5d
    print(f"5D basis vector {i+1} transforms to:", transformed)
```

```python
# Define the matrix
matrix = np.array([
    [2, 1, 3],
    [3, 1, 4],
    [5, 7, 12]
])

# Create a 3D plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the column vectors
origin = np.zeros(3)
for i in range(3):
    ax.quiver(*origin, *matrix[:, i], color=['r', 'g', 'b'][i], label=f'Column {i+1}')

# Plot the plane spanned by the vectors
xx, yy = np.meshgrid(range(10), range(10))
normal = np.cross(matrix[:, 0], matrix[:, 1])
z = (-normal[0] * xx - normal[1] * yy) / normal[2]
ax.plot_surface(xx, yy, z, alpha=0.2)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3x3 Matrix Columns as Vectors and the Plane They Span')

# Set axis limits
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 20])

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the matrix
matrix = np.array([
    [2, 1, 3],
    [3, 1, 4],
    [5, 7, 12]
])

# Create the figure
fig = go.Figure()

# Plot the column vectors
origin = np.zeros(3)
colors = ['red', 'green', 'blue']
for i in range(3):
    fig.add_trace(go.Scatter3d(
        x=[origin[0], matrix[0, i]],
        y=[origin[1], matrix[1, i]],
        z=[origin[2], matrix[2, i]],
        mode='lines',
        line=dict(color=colors[i], width=5),
        name=f'Column {i+1}'
    ))

# Plot the plane spanned by the vectors
xx, yy = np.meshgrid(range(10), range(10))
normal = np.cross(matrix[:, 0], matrix[:, 1])
z = (-normal[0] * xx - normal[1] * yy) / normal[2]

fig.add_trace(go.Surface(x=xx, y=yy, z=z, opacity=0.5, showscale=False))

# Set layout
fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        aspectmode='data'
    ),
		
    title='3x3 Matrix Columns as Vectors and the Plane They Span'
)

# Show the plot
fig.show()
```
